```python
import csv
import matplotlib.pyplot as plt

f = open('birthday_temp.csv',encoding = 'cp949')
d = csv.reader(f)
next(d)
min_temp=[1.0]*10
max_temp=[1.0]*10
index = 0
for row in d:
    split_data = row[0].split('.')
    if split_data[1] == '7' and split_data[2] == '19':
        row[3] = float(row[3])
        row[4] = float(row[4])
        min_temp[index]=row[3]
        max_temp[index]=row[4]
        index+=1
x = [0,1,2,5,10,15,20,25,30,35]
plt.plot(x,max_temp,color = 'violet')
plt.plot(x,min_temp,color = 'c')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('birthday_temp.csv',encoding = 'cp949')
d = csv.reader(f)
next(d)
jan_temp = []
aug_temp = []

for row in d:
    split_data = row[0].split('.')
    if row[4] != '':
        if split_data[1] == '1':
            jan_temp.append(float(row[4]))
        if split_data[1] == '8':
            aug_temp.append(float(row[4]))

plt.hist(jan_temp,bins=100,color='b',label = 'Jan')
plt.hist(aug_temp,bins=100,color='r',label = 'Aug')

plt.legend()
plt.show()
```
